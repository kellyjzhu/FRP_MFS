```python
import pandas as pd
import numpy as np
```

```python
fund_data = pd.read_csv("data0116.csv")
months = ["03","06","09","12"]
years = range(1990,2020)
dates = []
for month in months:
    for year in years:
        date = str(year)+month+ str(29)
        dates.append(date)
russell_stocks = pd.DataFrame()
for date in dates:
    new_data = pd.read_excel("russell" + str(date)+".xlsx",skiprows = 4)
    russell_stocks = pd.concat([russell_stocks, new_data], axis=0)
```

```python
fund_data = fund_data[pd.notnull(fund_data['calendar_qtr'])]
```

```python
fund_data
```

```python
def string_concat(name):
    data = str(name).split("Q")
    if data[1] == "1":
        return_value = data[0] + "-03-30"
    elif data[1] == "2":
        return_value = data[0] + "-06-30"
    elif data[1] == "3":
        return_value = data[0] + "-09-30"
    elif data[1] == "4":
        return_value = data[0] + "-12-31"

    return return_value

string_concat(fund_data['calendar_qtr'][0])
fund_data['calendar_date_qtr'] = fund_data['calendar_qtr'].apply(string_concat)
```
